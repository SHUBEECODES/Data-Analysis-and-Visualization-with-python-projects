# Key Performances

Most Player Of The Match Awards

```python
df.head()
```

```python
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y=count.index,palette='rainbow')
plt.title('Player of the match top 10')
```

Top 2 scorers

```python
df['top_scorer'].value_counts().head(2)
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
```

10 best bowling figures

```python
df.head()
```

```python
df['highbowlingfig']= df['best_bowling_figure'].apply(lambda x:x.split("--")[0])
df['highbowlingfig']= df['highbowlingfig'].astype(int)
topbowlers=df.groupby('best_bowling')['highbowlingfig'].sum().sort_values(ascending=False).head(10)
topbowlers
```

```python
topbowlers.plot(kind='barh')
```
